```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f_k(mu, k):
    # k is even
    half = k // 2
    numerator = math.factorial(k) * (mu**2 - 1)**half
    denominator = math.factorial(half) * (2**half)
    return numerator / denominator

# mu range
mu_vals = np.linspace(0.001, 0.999, 400)

# list of even k-values to plot
K_values = [2, 4, 6, 8] #, 10, 12]

plt.figure(figsize=(8,6))

for k in K_values:
    y_vals = [f_k(mu, k) for mu in mu_vals]
    plt.plot(mu_vals, np.abs(y_vals), label=f"k = {k}")

plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel(r"$\mu$")
plt.ylabel(r"$f_k(\mu)$")
plt.title(r"$f_k(\mu) = \frac{k!(\mu^2 - 1)^{k/2}}{(k/2)! 2^{k/2}}$ for even $k$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f_k(mu, k):
    # k is even
    half = k // 2
    numerator = math.factorial(k) * (mu**2 - 1)**half
    denominator = math.factorial(half) * (2**half)
    return numerator / denominator

# mu range
mu_vals = np.linspace(0.001, 0.999, 400)

# list of even k-values to plot
K_values = [2, 4, 6, 8] #, 10, 12]

plt.figure(figsize=(8,6))

for k in K_values:
    y_vals = [f_k(mu+1, k) for mu in mu_vals]
    plt.plot(mu_vals, np.abs(y_vals), label=f"k = {k}")

plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel(r"$\mu$")
plt.ylabel(r"$f_k(\mu)$")
plt.title(r"$f_k(\mu) = \frac{k!(\mu^2 - 1)^{k/2}}{(k/2)! 2^{k/2}}$ for even $k$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def solve_m(mu, u, tol=1e-12):
    """
    Return:
      - a float m if a unique solution exists,
      - the string 'any' if any m satisfies the equation,
      - the string 'none' if no solution exists.
    """
    # handle small numerical values
    if abs(u) < tol and abs(mu) < tol:
        # equation reduces to -1 = 0 -> impossible
        return 'none'
    if abs(u) < tol:
        # equation: mu^2 - 1 = 0
        return 'any' if abs(mu*mu - 1.0) < tol else 'none'
    if abs(mu) < tol:
        # equation: u^2 - 1 = 0
        return 'any' if abs(u*u - 1.0) < tol else 'none'
    # general case
    numerator = -(mu*mu + u*u - 1.0)
    denominator = 2.0 * u * mu
    m = numerator / denominator
    return m
```

```python
# Example
mu = 0.1
u  = 0.5
print(solve_m(mu, u))  # -> 0.4166666666666667
```
