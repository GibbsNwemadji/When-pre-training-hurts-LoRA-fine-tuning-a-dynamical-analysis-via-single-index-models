```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def V(g, mu):
    """
    V(g) = - (1 - mu)^2 * log(cosh(g))
    """
    return - (1.0 - mu)**2 * np.log(np.cosh(g))

# ---- g range ----
g_vals = np.linspace(-4.0, 4.0, 600)

# ---- mu values to compare ----
mu_vals = [0.0, 0.3, 0.7, 0.95]

plt.figure() #plt.figure(figsize=(9, 6))

for mu in mu_vals:
    V_vals = V(g_vals, mu)
    plt.plot(g_vals, V_vals, linewidth=2.0, label=rf"$\mu = {mu}$")

plt.xlabel(r"$g$")
plt.ylabel(r"$V(g)$")
plt.grid(True)
plt.tight_layout()
plt.legend(ncol=2, frameon=True)
plt.savefig("effective_linear_potential.pdf")
plt.show()
```
