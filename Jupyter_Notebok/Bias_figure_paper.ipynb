```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f_k_star(r, k_star: int):
    """
    f_{k*}(r) = k*! * (r-1)^(k*/2) / ( 2^(k*/2) * (k*/2)! )
    for even integer k*.
    """
    assert k_star % 2 == 0 and k_star >= 0
    m = k_star // 2
    const = math.factorial(k_star) / ( (2.0 ** m) * math.factorial(m) )
    return const * (r - 1.0) ** m

# ---- choose even k* values ----
k_stars = [2, 4, 6, 8, 10]   # edit as you like

# ---- r range (typically r >= 1 for this expression) ----
r_vals = np.linspace(0.001, 1.0, 400)

plt.figure() #

for k_star in k_stars:
    y = f_k_star(r_vals, k_star)
    plt.plot(r_vals, np.abs(y), linewidth=2.0, label=rf"$k^\star={k_star}$")

plt.xlabel(r"$r$")
plt.ylabel(r"$\sigma^{[r]}_0$")
plt.grid(True)
#plt.yscale("log")
plt.tight_layout()
plt.legend()
plt.savefig("bias_expression_vs_r.pdf")
plt.show()
```
